```python
import os
import numpy as np
import matplotlib.pyplot as plt
from digits_net_utils import read_config_file
%matplotlib inline
```

```python
config_file = os.path.join(os.getcwd(), "mnist_config.json")
```

```python
config = read_config_file(config_file)
```

```python
model_directory = config['model_directory'] + str(config['num_epochs'])
```

```python
losses_dict = np.load(os.path.join(model_directory, config['model_metrics'])).item()
```

```python
losses_dict.keys()
```

```python
plt.figure(figsize = (12, 8))
plt.xlabel('Epoch')
plt.ylabel('Avg. loss')
plt.title('Avg. loss vs. Epoch for Digit Recognition')
plt.plot(np.arange(1, len(losses_dict['train_loss']) + 1), losses_dict['train_loss'], 'b', label = "training loss")
plt.plot(np.arange(1, len(losses_dict['valid_loss']) + 1), losses_dict['valid_loss'], 'g', label = "validation loss")
plt.legend()
plt.show()
```
