```python
import os
import numpy as np
import matplotlib.pyplot as plt
from digits_net_utils import read_config_file
%matplotlib inline
```

```python
config_file = os.path.join(os.getcwd(), "mnist_config.json")
```

```python
config = read_config_file(config_file)
```

```python
train_losses = np.load(os.path.join(config['model_metrics'][0], config['model_metrics'][1]))
```

```python
plt.figure(figsize = (12, 8))
plt.xlabel('Epoch')
plt.ylabel('Avg. training loss')
plt.title('Avg. training loss vs. Epoch for Digit Recognition')
plt.plot(np.arange(1, len(train_losses) + 1), train_losses)
```
