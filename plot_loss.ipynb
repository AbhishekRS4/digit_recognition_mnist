```python
import os
import numpy as np
import matplotlib.pyplot as plt
from digits_net_utils import read_config_file
%matplotlib inline
```

```python
config_file = os.path.join(os.getcwd(), "mnist_config.json")
config = read_config_file(config_file)
```

```python
losses_vgg_file = os.path.join(os.getcwd(), os.path.join(config['model_directory'][0] + str(config['num_epochs']), config['model_metrics'][0]))
losses_res_file = os.path.join(os.getcwd(), os.path.join(config['model_directory'][1] + str(config['num_epochs']), config['model_metrics'][1]))
losses_pre_res_file = os.path.join(os.getcwd(), os.path.join(config['model_directory'][2] + str(config['num_epochs']), config['model_metrics'][2]))
```

```python
losses_dict_vgg = np.load(losses_vgg_file).item()
losses_dict_res = np.load(losses_res_file).item()
losses_dict_pre_res = np.load(losses_pre_res_file).item()
```

```python
plt.figure(figsize = (16, 12))
plt.xlabel('Epoch')
plt.ylabel('Avg. loss')
plt.title('Comparison of Avg. loss vs. Epoch for Digit Recognition')
plt.plot(np.arange(1, len(losses_dict_vgg['train_loss']) + 1), losses_dict_vgg['train_loss'], 'b', label = "vgg training loss")
plt.plot(np.arange(1, len(losses_dict_vgg['valid_loss']) + 1), losses_dict_vgg['valid_loss'], 'g', label = "vgg validation loss")
plt.plot(np.arange(1, len(losses_dict_res['train_loss']) + 1), losses_dict_res['train_loss'], 'c', label = "normal resnet training loss")
plt.plot(np.arange(1, len(losses_dict_res['valid_loss']) + 1), losses_dict_res['valid_loss'], 'r', label = "normal validation loss")
plt.plot(np.arange(1, len(losses_dict_pre_res['train_loss']) + 1), losses_dict_pre_res['train_loss'], 'm', label = "preactivation resnet training loss")
plt.plot(np.arange(1, len(losses_dict_pre_res['valid_loss']) + 1), losses_dict_pre_res['valid_loss'], 'k', label = "preactivation resnet validation loss")
plt.legend()
plt.show()
```
